```python
#converting the date column to datetime format
df["Date"]=pd.to_datetime(df["Date"],format="%d/%m/%Y")
```

```python
df["Day"] = df["Date"].dt.day
df["Week"]=df["Date"].dt.isocalendar().week
df["Month"] = df["Date"].dt.month
df["Year"] = df["Date"].dt.year
```

```python
df["IsHoliday_Same"] = df["IsHoliday_x"] == df["IsHoliday_y"]
```

```python
df["IsHoliday_Same"].value_counts()
```

```python
df.drop(["IsHoliday_x","IsHoliday_Same"],axis=1,inplace=True)
```

```python
df.rename(columns={"IsHoliday_y":"IsHoliday"},inplace=True)
df
```

Visualizing the Skewness and Outliers

```python
# function for box plot, hist plot, violin plot
def plot(df,column):
    plt.figure(figsize=(20,5))

    plt.subplot(1,3,1)
    sns.histplot(data= df1,x=column,kde=True,bins=30,color='salmon')
    plt.title(f'Distribution Plot for {column}')
    plt.show

    plt.subplot(1,3,2)
    sns.boxplot(data= df1,x=column,color='green')
    plt.title(f'Box Plot for {column}')
    plt.show

    plt.subplot(1,3,3)
    sns.violinplot(data=df1,x=column,color='indigo')
    plt.title(f'Violin Plot for {column}')
    plt.show
```

```python
df1.columns
```

```python
# plotting to check skew
for col in ['Store', 'Type', 'Size', 'Dept', 'Date', 'Weekly_Sales', 'Temperature',
       'Fuel_Price', 'MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4',
       'MarkDown5', 'CPI', 'Unemployment', 'IsHoliday', 'Day', 'Week', 'Month',
       'Year']:
    plot(df1,col)
```

# Feature Engineering

Converting the categorical data into numerical data

```python
df1["Type"]=df1["Type"].replace({"A":1,"B":2,"C":3})
```

```python
df1["IsHoliday"].unique()
```

```python
df1["IsHoliday"].value_counts()
```

```python
df1["IsHoliday"]=df1["IsHoliday"].apply(lambda x:1 if x==1 else 0)
```

```python
df1.drop("Date",axis=1,inplace=True)
```

```python
df1.info()
```

```python
df1["Day"]=df1["Day"].astype("int64")
df1["Week"]=df1["Week"].astype("int64")
df1["Month"]=df1["Month"].astype("int64")
df1["Year"]=df1["Year"].astype("int64")
```

```python
df1.info()
```

Creating Lag Features for Weekly Sales


```python
df1["Weekly_Sales_Lag1"]=df1["Weekly_Sales"].shift(1)
```

```python
df1['Weekly_Sales_Lag1'] = df1['Weekly_Sales_Lag1'].fillna(0)
```

```python
df1
```

Holiday Interaction Features

 Markdown and Holiday Interactions

```python
df1["Total_Markdowns"] = df1["MarkDown1"] + df1["MarkDown2"] + df1["MarkDown3"] + df1["MarkDown4"] + df1["MarkDown5"]
df1["Markdown_Holiday_Impact"] = df1["Total_Markdowns"] * df1["IsHoliday"]
```

```python
df1.drop("Total_Markdowns", axis=1, inplace=True)
```
